#### **Lecture/walk-through**


Some context before we begin this walk-through:

*   The DNA sequences 'TAG', 'TAA', and 'TGA' are [stop codons](https://en.wikipedia.org/wiki/Stop_codon).
*   These codons signal a 'stop' in the process of translating a corresponding mRNA sequence to a protein.


1. Using if/else statements, we can the gene HIST1H1B for the presence of a stop codon. Hopefully, we will not find any or our protein will be broken!


```python
if 'TAG' in histone_prot_genes['HIST1H1B']:
  print('There is a stop codon TAG')
elif 'TAA' in histone_prot_genes['HIST1H1B']:
  print('There is a stop codon TAA')
elif 'TGA' in histone_prot_genes['HIST1H1B']:
  print('There is a stop codon TGA')
else:
  print('There is no stop codon ATA')
```

This is somewhat worrisome! Our proteins will be broken! But wait, not every sequence of three letters equal to TAG can be interpreted as a stop codon because they needs to be one of the subsequent triplets of base pairs... E.g., for the string "AACGCAAAACCTGCTCTTTAGATT", we have "[AAC][GCA][AAA][CCT][GCT][CTT]**[TAG]**[ATT]".

Before looking into this, let us use a for loop to remove the "\n" and spaces from our gene sequences.

```python
for gene in histone_prot_genes:
  print(f"Fixing gene {gene}")
  histone_prot_genes[gene] = histone_prot_genes[gene].replace("\n", "")
  histone_prot_genes[gene] = histone_prot_genes[gene].replace(" ", "")
```

**Note about looping dictionaries**:

Looping dictionaries by default loops the keys of the dictionary.

`for value in histone_prot_genes.values():` can be used to loop the values instead of the keys of a dictionary.

`for key, value in histone_prot_genes.items():` can be used to loop both the keys and the values of a dictionary.

In `for gene in histone_prot_genes`, `gene` is a variable that receives one value at a time from the iterable `histone_prot_genes` (again, in this case, we iterated the keys of the dictionary).

**Back to our stop codon story...**

It turns out that not all base pairs of a gene are involved in the translation into a protein. For example, the [GenBank entry for this gene](https://www.ncbi.nlm.nih.gov/nuccore/NC_000006.12?report=genbank&from=27866792&to=27867588&strand=true) tells us the part of this sequence which is translated to form a protein starts at base pair 60 and ends at 740

image.png


Because Python indexes from 0 (instead of 1), the 60th character is at index 59. We can now use one triplet at a time and test if it is a stop codon:

```python
stop_codons = ['TAG', 'TAA', 'TGA']

prot_bp = histone_prot_genes['HIST1H1B']
for i in range(59, 740, 3):
  if prot_bp[i:i+3] in stop_codons:
    print(f"stop codon {prot_bp[i:i+3]} detected at position {i}!")
```

The stop codon is the last codon (triplet) of the sequences that base pairs that are meant to be used to translate the DNA into a protein. Yay! Our protein is safe!

Note that the function `range(start, stop, increment)` creates a sequence of numbers starting at `start`, finishing before `stop` (i.e., `stop` is not inclusive), and goes by increments `increment`.

### **Homework**
DNA strands serve as code for RNA strands. The transcription of DNA into RNA works by converting the bases as follows:
  * For every G in a DNA strand, the associated RNA base is C, and vice versa
  * For every A in a DNA strand, the associated RNA base is U
  * For every T in a DNA strand, the associated RNA base is A

Example: The corresponding RNA strand for 'ATTGCA' is 'UAACGU'  

Use a 'for' loop and if/else statements to construct the corresponding RNA sequence for the gene HIST1H1B.
Hints:
* If there are more than two conditions, use 'elif' after the first 'if'
* You can start with an empty string and add letters to it, like this:

rna_seq = ' '


rna_seq = rna_seq + 'A'
rna_seq = rna_seq + 'T'

This will give rna_seq = 'AT'

Note: You can also replace `x = x + y` by the shorter equivalent `x += y`

```python
rna_seq = ''
for char in histone_prot_genes['HIST1H1B']:
  if char == 'G':
    rna_seq += 'C'
  elif char =='A':
    rna_seq += 'U'
  elif char == 'T':
    rna_seq += 'A'
  elif char == 'C':
    rna_seq += 'G'

rna_seq
```

Can you acheive the same result using only the `.replace()` methods available with strings (e.g., `"a string".replace(" ", "_")`)

```python
rna_seq = histone_prot_genes['HIST1H1B']

# Note A needs to be replaced first.
# Can you see the issue if you were to invert the order?
rna_seq = rna_seq.replace("A", "U")
rna_seq = rna_seq.replace("T", "A")

# Can you see why we need a temporary symbol "*"?
rna_seq = rna_seq.replace("G", "*")
rna_seq = rna_seq.replace("C", "G")
rna_seq = rna_seq.replace("*", "C")

rna_seq
```
